### Data Structure

In the data structure, we look out for 2 things: if it is **ordered** and if it is **mutable**
- **List** is an ordered collection of items. Lists can be mutable, meaning you can change their content. [1, "cat", true, 1, 2]
- **Sets** are mutable and unordered data structures of unique items. {1, 2, 3, 4, 5}
- **Tuples** are very much like lists. They are ordered but immutable. (10.0, 20.0)
- **Dictionary** - they are values: key pair mutable data structure that can be ordered. {"name": "john", "age": 9}

### Common List Methods
- **append(item)**: Adds an item to the end of the list
- **extend(iterable)**: Adds all elements of an iterable (like another list) to the end of the list.
- **insert(index, item)**: inserts an item to a given position
- **remove(item)**: removes the first occurrence of an item
- **pop(index)**: removes and returns the item at a given position(default is the last item).
- **sort()**: to sort in an ascending order
- **reverse()**: to reverse the list
- **count(item)**: returns the number of occurrences of an item
- **index(item)**: returns the index of the first occurrence of an item

```python
list_1 = [1, 2, 4, 3, {"name": "olu"}, True, -5, "friyah", 4 ]
print(list_1)
```

### Common Tuples Methods
- **index(item)**: returns the index of the first occurrence of the item
- **count(item)**: returns the number of times an item appears in the tuple

### Common Sets Methods
- **add(item)**: adds an set to the set.
- **remove(item)**: removes a specific element from the set.
- **discard(item)**: removes a specific element if it exists, without raising an error if it doesn't.
- **pop()**: removes and returns a random element.
- **clear()**: removes all elements from the set.
- **union(set2)**: returns a set containing all elements from both sets.
- **intersection(set2)**: returns a set containing elements found in both sets
- **difference(set2)**: returns a set containing elements in the first set but not in the second

### Common Dictionary Methods
- **get(key)**: returns the value of a specified key
- **keys()**: returns all the keys in the dictionary.
- **values()**: returns all the values in the dictionary.
- **items()**: returns all the key-value pairs as tuples.
- **update(dict2)**: updates the dictionary with the key-value pairs from another dictionary.
- **pop(key)**: removes and returns the value for the specified key.
- **clear()**: removes all key-value pairs from the dictionary

### LOOP

**Loop** are used to avoid writing repetitive codes. They allow us to repeat actions on conditions.

#### For Loop
- **For loop** is used to iterate over a sequence (such as a list, tuple, dictionary, set or string) or other iterable object.
-  **A for loop** is used to repeat actions for a known number of items. Think of it like going through a list or checklist, doing something for each item.
-  **For loop** is useful when you know how many times you need to repeat an action (like a list of items to go through).

```python
for r in range(1,10):
  print(r)
```

**The positioning** of print, increment, and break in the loop controls which value are printed and when the loop stop

```python
# The use of a break after the print statement
fruits = ["apple", "banana", "orange", "avocado", "cherry", "mango", "banana"]
for x in fruits:
    print(x)
    if x == "banana":    
        break
```

```python
# The use of break before the print statement
fruits = ["apple", "banana", "orange", "avocado", "cherry", "mango", "banana"]
for x in fruits:
    if x == "banana":
       break
    print(x)
```

```python
# The use of continue fruits = ["apple", "banana", "orange", "avocado", "cherry", "mango", "banana"]
for x in fruits:
    if x == "banana":    
        continue
    print(x)   
```

```python
# Creating another list that has the letter a in the fruit's name using traditional method
fruits = ["apple", "banana", "orange", "avocado", "cherry", "mango", "banana"]
afruits = ["Date"]

for x in fruits:
    if "a" in x:
        afruits.append(x)
print(afruits)
```

```python
# Creating another list that has the letter a in the fruit's name using list comprehension method
fruits = ["apple", "banana", "orange", "avocado", "cherry", "mango", "banana"]
afruits = ["Date"] + [x for x in fruits if "a" in x]
print(afruits)
```

#### While Loop
- **A while loop** repeatedly execute a block of code as long as the condition is true
- Setting a task thats repeats until something happens

```python
i = 1
while i < 6:
   print(i)
   i += 1  
```

```python
# While loop with the break statement
# If you want 3 to be included before its breaks
i = 1
while i < 6:
    print(i)
    if i == 3:
        break
    i += 1    
```

```python
# While loop with the break statement
# If you dont want 3 to be included before its breaks
i = 1
while i < 6:
    print(i)
    i += 1 
    if i == 3:
        break
```

```python
# While loop with continue statement
i = 0 
while i < 6:
    i += 1
    if i == 3:
        continue
    print(i) 
```

```python
# ATM Example
balance = 1000
withdraw = True
while balance > 0 and withdraw:
    print("Input amount to continue or (0 to stop)")
    amount = float(input("Enter amount to withdraw: "))

    if amount == 0:
        print("Existing withdrawal process")
        withdraw = False
    elif amount > balance:
        print("Insuffiecient balance, try smaller amount")

    else :
        balance -= amount
    print(f"You have successfully withdrawn {amount}. Your balance is {balance}")

if balance == 0:
    print("You have withdrawn all")
```
