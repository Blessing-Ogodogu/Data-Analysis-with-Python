<p style="color: black;">Student performance is influenced by a variety of factors, including academic habits, socioeconomic background, parental involvement, and lifestyle choices. Understanding the key drivers behind student success can enable educators, policymakers, and guardians to tailor interventions that enhance academic outcomes. This dataset provides a comprehensive view of student performance-related factors, such as study habits, family dynamics, socioeconomic status, and health behaviors. By exploring these dimensions, this analysis aims to identify significant predictors of academic achievement and provide data-driven insights that may guide strategies for improved educational outcomes.

</div>

<p style="color: black;">
Despite efforts to enhance educational standards, many students struggle to achieve satisfactory academic performance. Numerous factors—ranging from family background and access to resources to individual lifestyle choices—affect students' academic success. However, it is challenging to pinpoint which of these factors has the most significant impact on student outcomes. Without clear data on these influences, it is difficult to design targeted interventions that address the root causes of poor performance. This analysis seeks to identify and evaluate the most influential factors impacting students' exam scores, thereby offering a clearer understanding of the multifaceted nature of student performance.
</p>

<div style="text-align: left; font-size: 20px; font-weight: 700; color: purple;">
   4.3 Attendance AND Exam_Score ANALYSIS WITH A LINE CHART
</div>

```python
fd = my_data.groupby("Attendance")["Exam_Score"].sum().sort_values(ascending = False)

print(fd)

fd.head(5)
```

```python
sns.lineplot(x="Attendance", y="Exam_Score", data=my_data, color="purple")
plt.title("Attendance vs Exam Score")
plt.xlabel("Attendance")
plt.ylabel("Exam Score")
```

<div style="text-align: left; font-size: 20px; font-weight: 700; color: purple;">
   4.3 GENDER ANALYSIS WITH ATTENDANCE AND EXAM SCORE AGGREGATE
</div>

```python
my_data.groupby("Gender").agg({"Attendance" : "sum", "Exam_Score" : "sum"}).sort_values(by = "Exam_Score", ascending = False)
```

```python
# Pie Chart with Matplotlib
pt = my_data.groupby("Gender")["Attendance"].sum()
colors = ["indigo", "purple"]
plt.pie(pt, labels=pt.index, autopct= "%1.1f%%", startangle=90, colors=colors)
plt.title("Gender vs Attendance")
```

```python
# Pie Chart with Matplotlib
pt = my_data.groupby("Gender")["Exam_Score"].average()
colors = ["indigo", "purple"]
plt.pie(pt, labels=pt.index, autopct= "%1.1f%%", startangle=90, colors=colors)
plt.title("Gender vs Exam Score")
```

<span style="text-align: left; font-size: 15px; font-weight: 700; color: purple;">
  OBSERVATION: 
</span><span style="color: black;">Males have a higher total attendance (295,285) compared to females (215,088), suggesting that male students, on average, may have higher attendance.Similarly, the total Exam_Score for males is higher (247,965) than for females (180,969)
 </span>

<div style="text-align: left; font-size: 20px; font-weight: 700; color: purple;">
   5.1 RECOMMENDATION
</div>

- Higher hours studied and better attendance tend to correlate positively with higher exam scores, indicating that consistent study routines and class participation play a significant role in performance.

- Factors such as parental involvement and parental education level appear to impact exam scores, suggesting that support from family is crucial for student success.

- Family income and access to resources show some influence on exam scores, highlighting that students from better socioeconomic backgrounds may have more academic advantages.

- Male gender tend to have higher attenadnce and also higher exams score than the female gender
