### Visualization

- **Pandas:** Built on matplotlib, simplifies quick plotting directly from DataFrame
- **Seaborn:** Built on matplotlib, simplifies complex statistical visualization and improves aesthetics of the plots.
- **Matplotlib:** The foundation of both pandas and seaborn visualization. Its provides the core plotting functions, but more flexibility at the cost of the increased complexity.

##### **Where to use each library**

- **Pandas:** When you need quick, basic visualization directly from the DataFrame
- **Seaborn:** When you want aesthetically pleasing plots with less code and when you are focusing on statistical relationships in the data. 
- **Matplotlib:** When you need the control and customization of plots or when creating more complex visualizations.

##### **Most commonly used chart**

- **Bar Chart**: Categorical comparison using rectangular bars.
   - **When to use**: Comparing different groups or categories (e.g sales by region, products by popularity)
   - **Used for**: Discrete categories. When you have a few categories.
- **Histogram**: Distribution of a numeric variable using bins.
   - **When to use**: To observe the frequency distribution of data(e.g age distribution, ranges)
   - **Used for**: Understanding the spread or skewness of data.
- **Line Chart**: Used for time series data.
   - **When to use**: To how trends or changes over time ( e.g stock prices over months)
   - **Used for**: Tracking changes over a continuous interval (like time).
- **Scatter Plots**: Points plotted on two axes to display the relationship between two variables.
   - **When to use**: To show a correlation or relationship between 2 numeric variables._(e.g sales vs profits)
   - **Used for**: Identifying relationships, clusters or outliers.
- **Pie Chart**: Circular charts divided into slices to represent proportions.
   - **When to use**: To show parts of a whole (e.g market shares, budget allocations)
   - **Used for**: Simple proportional comparison, avoid too many categories.
- **Boxplot**: Display distributions, quartiles, and outliers of data. 
   - **When to use**: To compare distributions across categories(e.g profit across different product lines)
   - **Used for**: Visualizing spreads outliers, and comparing distributions particularly with statistical data.

## Outliers

- Outliers are data points thats deviate significantly from the rest of the dataset. They can be much higher or lower than most of the values.
- Example, in a dataset of salaries, if salaries are between 30, 000 and 50,000 but one salary is bout $500,000, this an outlier

### Why are outliers important?
- **Influence on Results:** Outliers can distort statistical measures(like the mean) and affect machine learning models by leading to biased or inaccurate results.
- **Data Quality:** Outliers might indicate errors in data collection, incorrect data entry or rare or important cases.

### How to detect Outliers
- **Box plots**: Outliers are often represented as points outside the whiskers of a boxplot.
- **Histograms**: Extremely high or low values will appear isolated from the rest of the data.
- **Scatter plots**: For bivariate data, points that are far from the general cluster can be outliers.
- **IQR (Interquartile range)**: Calculate the IQR as the difference between the 75th percentile (Q3) and the 25th percentile (Q1). Outliers are typically data points that fall below Q1 - 1.5 * IQR or above Q3 + 1.5 *IQR

### Handling Outliers 
- **Keep them**: In some cases, outliers are valid data points that represent rare but real phenomena. For example a company with unusually high sale  could represent a valid or valuable case.
- **Remove them**: If outliers results from errors or noise, removing them can lead to more accurate analysis.
- **Cap them**: In some cases, you can 'cap' them by replacing extremely values with m.aximum and minimum threhold(winsorizing).
- **Transform them**: Use log transformations or other techniques to reduce the influence of outliers while keeping them in the dataset.


### Detecting and Handling Outliers in Python
- Boxplot Example:

```python
sd.boxplot(column = 'Profit')
plt.title('Box plot for Profit')
```

- **IQR Example**

```python
Q1 = sd['Profit'].quantile(0.25)
Q3 = sd['Profit'].quantile(0.75)


IQR = Q3 - Q1
print(Q1)
print(Q3)
print(IQR)
```

```python
#Filter outliers
outliers = sd[(sd['Profit'] < (Q1 - 1.5 * IQR)) | (sd['Profit'] > (Q3 + 1.5 * IQR))]
```

```python
outliers.shape
```

```python
outliers
```

```python
sd[~(sd['Profit'] < (Q1 - 1.5 * IQR)) | (sd['Profit'] > (Q3 + 1.5 * IQR))]
```

### Why and when to handle 
- **When outliers distort models**: Machine learning models, especially those based on regression or distance metric(like KNN) can be highly sensitive to outliers.
- **When outliers are errors**: If outliers are due to data entry mistakes or sensor malfuntions removing or correcting them is important for data integrity.
